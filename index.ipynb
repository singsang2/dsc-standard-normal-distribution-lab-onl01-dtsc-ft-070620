## Briefly explain what each value represents in this data set

```python
# Your answer here
df_stat = df['0'].describe()
```

## Define the interval bounds which contain 99% of the observations   

> **Hint**: Recall the empirical rule related to $3\sigma$.

```python
# Perform any calculations necessary here
# 68, 95, 99
df_stat['mean']-3*df_stat['std'], df_stat['mean']+3*df_stat['std']
```

```python
# Write your answer here 
mu = df_stat['mean']
std = df_stat['std']
```

## Compute and interpret the z-score for a tree yielding 35 pounds of apples

```python
# Calculate z
z_score = (35 - mu)/std
```

```python
# Interpret the result
print(f'35lb is {z_score} standard deviation away from the mean to the left')
```

## Suppose a tree has a z-score of 1.85. Interpret this z-score. What is the yield of this tree?

```python
# Interpret the z score
z_score = 1.85
print(f'This tree is {z_score} standard deviation away from the mean to the left')
```

```python
# Calculate yield
yield_ = z_score*std + mu
```

```python
# What is the yield ?
print('The yield is {}lb.'.format(yield_))
```

##  Convert each tree’s yield to a z-score so the new variable is the “z-score for weight”

The units are still the apple trees. For the data set of all z-scores:

* What is the shape? 
* The mean? 
* The standard deviation?

```python
z_scores = [(x-mu)/std for x in df['0']]
sns.distplot(z_scores)
```

```python
mu_z = np.array(z_scores).mean()
std_z = np.array(z_scores).std()
print('mu_z = {}\nstd_z = {}'.format(mu_z, std_z))
```
